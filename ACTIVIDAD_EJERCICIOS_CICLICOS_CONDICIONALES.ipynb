```python
valor_boleta = float(input("Ingrese el valor base de la boleta: "))

total_recaudado = 0
total_descuento = 0

for i in range(3):
    estrato = int(input(f"\nPersona {i + 1} - Ingrese el estrato (1 o 2): "))
    edad = int(input("Ingrese la edad: "))


    if estrato == 1:
        descuento = 0.20 if edad < 18 else 0.15
    elif estrato == 2:
        descuento = 0.10 if edad < 18 else 0.05
    else:
        print("Estrato no válido. No se aplicará descuento.")
        descuento = 0

    descuento_aplicado = valor_boleta * descuento
    pago = valor_boleta - descuento_aplicado

    total_recaudado += pago
    total_descuento += descuento_aplicado

    print(f"Descuento aplicado: ${descuento_aplicado:.2f}")
    print(f"Valor a pagar: ${pago:.2f}")

print("\nResumen final:")
print(f"Total recaudado: ${total_recaudado:.2f}")
print(f"Total descontado: ${total_descuento:.2f}")
```
